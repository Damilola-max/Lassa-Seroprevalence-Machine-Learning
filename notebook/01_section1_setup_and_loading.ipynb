## CELL 1.2: CONFIGURE ENVIRONMENT FOR REPRODUCIBILITY

### What We Do
Set random seeds and configure display options to ensure reproducibility.

### Why Set Random Seeds?

#### Problem Without Seeds
```
Run 1: Model accuracy = 0.87
Run 2: Model accuracy = 0.82  ← Different result!
Run 3: Model accuracy = 0.85  ← Different again!
```
**Why?** Random initialization, data shuffling, etc.

#### Solution: Set Random Seed
```
np.random.seed(42)  → Initialize NumPy with seed 42
random.seed(42)     → Initialize Python random with seed 42

Run 1: Model accuracy = 0.87
Run 2: Model accuracy = 0.87  ← SAME!
Run 3: Model accuracy = 0.87  ← SAME!
```
**Why 42?** Arbitrary choice, but famous in computer science.

### Why Configure Display Options?
- **Show all columns**: See complete DataFrames
- **Full width**: No truncation
- **Float precision**: Show accurate numbers (6 decimals)

### Why Configure Visualization?

- **DPI**: Crisp, clear plots

## CELL 1.4: LOAD DATA FROM GITHUB

### What We Do
Load LASV_Master_Data!.csv using smart fallback logic.

### Why Load from GitHub?
- **Version Control**: Track data changes
- **Reproducibility**: Anyone can load same data
- **Transparency**: Data source documented in code
- **Collaboration**: Easy to share data source

### Load Strategy
```
If file exists locally
  ├─ Load from local file (FAST)
  └─ Skip GitHub (offline friendly)
Else
  ├─ Try downloading from GitHub (SLOW)
  ├─ Save to local for future use
  └─ Handle errors gracefully
```

````python
# ============================================================================
# CELL 1.4: LOAD DATA FROM GITHUB
# ============================================================================
# Purpose: Load LASV_Master_Data!.csv with smart fallback
# Strategy: Try local first (fast), then GitHub (slow), then error
# Runtime: ~2-10 seconds (depends on internet)
# ============================================================================

import urllib.request

print("\n" + "="*100)
print("CELL 1.4: LOAD DATA FROM GITHUB")
print("="*100)

print("\nLoading data...")

# Define data file path
data_path = 'data/LASV_Master_Data!.csv'

# ============================================================================
# STRATEGY 1: TRY LOADING FROM LOCAL FILE (FAST, OFFLINE)
# ============================================================================

if os.path.exists(data_path):
    # File exists locally
    print(f"✓ Local file found: {data_path}")
    df_raw = pd.read_csv(data_path)
    print(f"✓ Loaded from local file")
    print(f"  Speed: FAST (no internet needed)")
    print(f"  Status: Offline mode enabled")

else:
    # File not found locally, try GitHub
    print(f"⚠ Local file not found at {data_path}")
    print(f"Attempting GitHub download...\n")
    
    try:
        # ============================================================================
        # CORRECT GitHub raw content URL
        # ============================================================================
        github_url = "https://raw.githubusercontent.com/Damilola-max/Lassa-Seroprevalence-Machine-Learning-/main/data/LASV_Master_Data.csv"
        
        print(f"Repository: Damilola-max/Lassa-Seroprevalence-Machine-Learning-")
        print(f"Branch: main")
        print(f"File: files/data/LASV_Master_Data!.csv")
        print(f"URL: {github_url[:70]}...\n")
        
        # ============================================================================
        # Download file using urllib (more reliable than pd.read_csv for URLs)
        # ============================================================================
        print(f"✓ Downloading from GitHub...")
        urllib.request.urlretrieve(github_url, data_path)
        print(f"✓ Download complete!\n")
        
        # Load the downloaded file
        df_raw = pd.read_csv(data_path)
        print(f"✓ File loaded successfully from: {data_path}")
        print(f"  Speed: SLOW (first run, needs internet)")
        print(f"  Next run: Will use local file (fast)")
        
    except urllib.error.HTTPError as e:
        # GitHub download failed - specific HTTP error
        print(f"✗ HTTP Error {e.code}: {e.reason}")
        print(f"\nTROUBLESHOOTING:")
        print(f"  1. Check internet connection")
        print(f"  2. Verify repository is PUBLIC")
        print(f"  3. Verify file exists at:")
        print(f"     https://github.com/Damilola-max/Lassa-Seroprevalence-Machine-Learning-/tree/main/files/data")
        print(f"  4. Download file manually:")
        print(f"     - Visit: https://github.com/Damilola-max/Lassa-Seroprevalence-Machine-Learning-")
        print(f"     - Navigate to: files/data folder")
        print(f"     - Download: LASV_Master_Data!.csv")
        print(f"     - Place in: data/ folder")
        print(f"     - Re-run this cell")
        raise
        
    except Exception as e:
        # Other download errors
        print(f"✗ Error downloading: {str(e)}")
        print(f"\nTROUBLESHOOTING:")
        print(f"  1. Check internet connection")
        print(f"  2. Check if file path is correct")
        print(f"  3. Download manually and place in: data/LASV_Master_Data!.csv")
        raise

# ============================================================================
# CREATE WORKING COPY
# ============================================================================
# Why separate copy?
# - df_raw = original (never modify)
# - df = working copy (for processing)
# This prevents accidental data loss

df = df_raw.copy()
print(f"\n✓ Working copy created (df)")
print(f"  - df_raw: Original data (unchanged)")
print(f"  - df: Working copy (for processing)")

# ============================================================================
# VERIFY DATA LOADED
# ============================================================================

print(f"\n✓ Data loaded successfully")
print(f"  Shape: {df_raw.shape[0]} rows × {df_raw.shape[1]} columns")
print(f"  Memory: {df_raw.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"  Size: Small (easy to work with)")

# ============================================================================
# Verify files in data directory
# ============================================================================
print(f"\n📁 Files in data/ directory:")
if os.path.exists('data'):
    for f in os.listdir('data'):
        if f.endswith('.csv'):
            size = os.path.getsize(f'data/{f}') / 1024  # KB
            print(f"  ✓ {f}: {size:.1f} KB")

print("\n" + "="*100)
````

## CELL 1.6: DATA QUALITY ASSESSMENT

### What We Do
Check for data quality issues.

### Critical Finding: CLASS IMBALANCE
```
Negative (healthy):  241 patients (96.4%)
Positive (infected):   9 patients (3.6%)

Ratio: 1:26.8 (SEVERE IMBALANCE)
```
**Impact:** Standard models are biased toward "negative"
**Solution:** SMOTE + weighted loss + threshold optimization

````python
# ============================================================================
# CELL 1.6: DATA QUALITY ASSESSMENT
# ============================================================================
# Purpose: Check for data quality issues
# Runtime: ~2 seconds
# Output: Data quality report
# ============================================================================

print("\n" + "="*100)
print("CELL 1.6: DATA QUALITY ASSESSMENT")
print("="*100)

# ============================================================================
# 1. MISSING VALUES ANALYSIS
# ============================================================================

print("\n1. MISSING VALUES:")
print("-" * 100)

missing_total = df.isnull().sum().sum()
missing_pct = (missing_total / (df.shape[0] * df.shape[1]) * 100)

print(f"  Total missing cells: {missing_total}")
print(f"  As percentage: {missing_pct:.2f}%")
print(f"  Status: {'✓ MINIMAL' if missing_pct < 5 else '⚠ SIGNIFICANT'}")
print(f"  Action: Will use median/mode imputation")

# Show columns with missing values
missing_by_col = df.isnull().sum()
missing_by_col = missing_by_col[missing_by_col > 0].sort_values(ascending=False)

if len(missing_by_col) > 0:
    print(f"\n  Columns with missing values ({len(missing_by_col)}):")
    for col, count in missing_by_col.head(10).items():
        pct = count / len(df) * 100
        print(f"    - {col:<40}: {count:>3} ({pct:>5.1f}%)")
else:
    print(f"  ✓ No missing values in any column")

# ============================================================================
# 2. DUPLICATE ROWS
# ============================================================================

print("\n2. DUPLICATE ROWS:")
print("-" * 100)

dup_count = df.duplicated().sum()
print(f"  Total duplicates: {dup_count}")
print(f"  Status: {'✓ GOOD' if dup_count == 0 else '⚠ ACTION NEEDED'}")

if dup_count > 0:
    print(f"  Action: Will remove {dup_count} duplicate rows")
else:
    print(f"  No duplicates found")

# ============================================================================
# 3. TARGET VARIABLE ANALYSIS (CRITICAL)
# ============================================================================

print("\n3. TARGET VARIABLE ANALYSIS (CRITICAL):")
print("-" * 100)

target = 'lab_results.PCR_Results'

if target in df.columns:
    print(f"  Target: {target}")
    print(f"  Type: {df[target].dtype}")
    
    target_dist = df[target].value_counts()
    print(f"\n  Class distribution:")
    for val, count in target_dist.items():
        pct = count / len(df) * 100
        bar_length = int(pct / 2)
        bar = '█' * bar_length
        print(f"    {val:<20}: {count:>3} ({pct:>6.2f}%) {bar}")
    
    if len(target_dist) == 2:
        positive_count = target_dist.get('320Kb (Positive)', 0)
        negative_count = target_dist.get('No Kb (Negative)', 0)
        
        imbalance_ratio = negative_count / max(positive_count, 1)
        
        print(f"\n  ⚠ CLASS IMBALANCE RATIO: 1:{imbalance_ratio:.1f}")
        
        if imbalance_ratio > 10:
            print(f"  STATUS: SEVERE IMBALANCE (> 1:10)")
        elif imbalance_ratio > 3:
            print(f"  STATUS: MODERATE IMBALANCE")
        else:
            print(f"  STATUS: BALANCED")
        
        print(f"\n  IMPLICATIONS:")
        print(f"  • Default models are biased toward majority class")
        print(f"  • Accuracy metric is misleading (96.4% always predicting negative)")
        print(f"  • SOLUTIONS REQUIRED:")
        print(f"    1. SMOTE: Create synthetic positive samples")
        print(f"    2. Weighted loss: Penalize misclassifying positives")
        print(f"    3. Threshold optimization: Adjust decision boundary")
        print(f"    4. F1/ROC-AUC metrics: Use instead of accuracy")
else:
    print(f"  ✗ Target column not found!")

# ============================================================================
# 4. DATA TYPE SUMMARY
# ============================================================================

print("\n4. DATA TYPE SUMMARY:")
print("-" * 100)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print(f"  Numerical columns: {len(numerical_cols)}")
print(f"    → Need scaling, normalization")
print(f"  Categorical columns: {len(categorical_cols)}")
print(f"    → Need encoding (one-hot, label)")
print(f"  Total: {len(numerical_cols) + len(categorical_cols)}")

print("\n" + "="*100)
print("✓ Data quality assessment complete")
print("="*100)
````
