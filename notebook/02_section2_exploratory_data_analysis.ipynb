## CELL 2.2: NUMERICAL FEATURES ANALYSIS

### What We Do
Analyze distributions and statistics of numerical features.

### Numerical Features (3 total)
1. **Age**: Patient age in years (1-110)
2. **lab_results.IgM OD_Values**: IgM antibody optical density (0-2.83)
3. **lab_results.IgG OD_Values**: IgG antibody optical density (0-0.87)

### Statistics Included
- Count, Mean, Std Dev
- Min, Max, Quartiles
- Distribution shape
- Outlier detection

### Why This Matters
- Understand feature ranges and distributions
- Identify outliers and unusual values
- Plan scaling strategy
- Detect potential data quality issues

```python
print('\n' + '='*100)
print('CELL 2.2: NUMERICAL FEATURES ANALYSIS')
print('='*100)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f'\nFound {len(numerical_cols)} numerical features:')
for i, col in enumerate(numerical_cols, 1):
    print(f'  {i}. {col}')

print('\n' + '-'*100)
print('DETAILED STATISTICS FOR EACH FEATURE')
print('-'*100)

for col in numerical_cols:
    print(f'\n📊 FEATURE: {col}')
    print(f'   Count:      {df[col].count()}')
    print(f'   Mean:       {df[col].mean():.6f}')
    print(f'   Std Dev:    {df[col].std():.6f}')
    print(f'   Min:        {df[col].min():.6f}')
    print(f'   Q1 (25%):   {df[col].quantile(0.25):.6f}')
    print(f'   Median:     {df[col].quantile(0.50):.6f}')
    print(f'   Q3 (75%):   {df[col].quantile(0.75):.6f}')
    print(f'   Max:        {df[col].max():.6f}')
    
    iqr = df[col].quantile(0.75) - df[col].quantile(0.25)
    lower_bound = df[col].quantile(0.25) - 1.5 * iqr
    upper_bound = df[col].quantile(0.75) + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()
    
    print(f'   IQR:        {iqr:.6f}')
    print(f'   Outliers:   {outliers} ({round(outliers/len(df)*100, 2)}%)')
    print(f'   Skewness:   {df[col].skew():.6f}')
    print(f'   Kurtosis:   {df[col].kurtosis():.6f}')

print('\n' + '='*100)
print('✓ CELL 2.2 COMPLETE: NUMERICAL FEATURES ANALYZED')
print('='*100)
```

## CELL 2.4: TARGET VARIABLE DEEP DIVE

### What We Do
Comprehensive analysis of the target variable.

### Target Variable
- **Name**: lab_results.PCR_Results
- **Type**: Binary Classification (2 classes)
- **Classes**: 
  - Negative (No Kb): 247 samples (98.80%)
  - Positive (320Kb): 3 samples (1.20%)

### Critical Finding
**SEVERE CLASS IMBALANCE: 1:82.3 ratio**

### Analysis Includes
- Class distribution
- Imbalance ratio
- Statistical impact
- Modeling implications

### Why This Matters
- Standard models will predict 98.8% negative always
- Accuracy metric is misleading
- Need special handling (SMOTE, class weights)
- Must use F1-Score instead of Accuracy

```python
print('\n' + '='*100)
print('CELL 2.4: TARGET VARIABLE DEEP DIVE')
print('='*100)

target = 'lab_results.PCR_Results'
print(f'\nTarget Variable: {target}')
print(f'Data Type: {df[target].dtype}')
print(f'Total Samples: {len(df)}')

print('\n' + '-'*100)
print('CLASS DISTRIBUTION')
print('-'*100)

target_dist = df[target].value_counts()
target_pct = df[target].value_counts(normalize=True) * 100

for val, count in target_dist.items():
    pct = target_pct[val]
    bar_length = int(pct / 2)
    bar = '█' * bar_length
    print(f'\n{val}')
    print(f'  Count:       {count}')
    print(f'  Percentage:  {pct:.2f}%')
    print(f'  Bar:         {bar}')

positive = target_dist.get('320Kb (Positive)', 0)
negative = target_dist.get('No Kb (Negative)', 0)

print('\n' + '-'*100)
print('IMBALANCE ANALYSIS')
print('-'*100)

if positive > 0:
    imbalance_ratio = negative / positive
    print(f'\nPositive samples:  {positive}')
    print(f'Negative samples:  {negative}')
    print(f'Imbalance ratio:   1:{imbalance_ratio:.1f}')
    
    if imbalance_ratio > 50:
        severity = 'CRITICAL'
    elif imbalance_ratio > 20:
        severity = 'SEVERE'
    elif imbalance_ratio > 10:
        severity = 'MODERATE'
    else:
        severity = 'MILD'
    
    print(f'Severity:          {severity}')

print('\n' + '-'*100)
print('MODELING IMPLICATIONS')
print('-'*100)

print('\n⚠ PROBLEMS WITH STANDARD MODELS:')
print('  1. Bias: Model learns to predict negative always')
print('  2. Accuracy: Misleading metric (98.8% by predicting negative)')
print('  3. Recall: Very low for positive class')
print('  4. Precision: Can be high but low absolute positives detected')

print('\n✓ REQUIRED SOLUTIONS:')
print('  1. SMOTE: Generate synthetic positive samples')
print('  2. Class Weights: Penalize wrong predictions more')
print('  3. Metrics: Use F1-Score, ROC-AUC, not Accuracy')
print('  4. Stratification: Maintain ratio in train-test split')
print('  5. Threshold: Optimize decision boundary')

print('\n' + '='*100)
print('✓ CELL 2.4 COMPLETE: TARGET VARIABLE ANALYZED')
print('='*100)
```

## CELL 2.7: ENCODING STRATEGY PLANNING

### What We Do
Plan how to encode categorical features for modeling.

### Encoding Methods
1. **Label Encoding**: Binary features (2 values) → 0/1
2. **One-Hot Encoding**: Low cardinality (3-5 values) → binary columns
3. **Target Encoding**: High cardinality (>5 values) → target mean

### For This Dataset
- Most features are binary (Yes/No)
- Recommend: One-Hot Encoding (interpretable)
- Alternative: Label Encoding (simpler)

### Why This Matters
- ML models need numerical input
- Choice affects model interpretability
- Impact on feature space size
- Influences model performance

```python
print('\n' + '='*100)
print('CELL 2.7: ENCODING STRATEGY PLANNING')
print('='*100)

categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
target = 'lab_results.PCR_Results'
categorical_cols = [col for col in categorical_cols if col != target]

print(f'\nCategorical Features to Encode: {len(categorical_cols)}')

print('\n' + '-'*100)
print('CARDINALITY ASSESSMENT')
print('-'*100)

encoding_strategy = []
binary_count = 0
lowcard_count = 0
highcard_count = 0

for col in categorical_cols:
    unique = df[col].nunique()
    
    if unique == 2:
        strategy = 'Label Encoding'
        binary_count += 1
    elif unique <= 5:
        strategy = 'One-Hot Encoding'
        lowcard_count += 1
    else:
        strategy = 'Target Encoding'
        highcard_count += 1
    
    encoding_strategy.append({
        'Feature': col,
        'Unique': unique,
        'Strategy': strategy
    })

print('\nTop 15 Features with Strategy:')
for i, item in enumerate(encoding_strategy[:15], 1):
    print(f'{i:2}. {item["Feature"]:<40} | Unique: {item["Unique"]:>2} | {item["Strategy"]}')

print('\n' + '-'*100)
print('ENCODING SUMMARY')
print('-'*100)

print(f'\nLabel Encoding (Binary):    {binary_count} features')
print(f'  Formula: feature_value in [0, 1]')
print(f'  Example: Yes→1, No→0')
print(f'  Pros: Simple, interpretable, fewer features')

print(f'\nOne-Hot Encoding:           {lowcard_count} features')
print(f'  Formula: Create n_categories binary columns')
print(f'  Example: Male→[1,0], Female→[0,1]')
print(f'  Pros: Preserves all information, interpretable')

print(f'\nTarget Encoding:            {highcard_count} features')
print(f'  Formula: Replace with target mean')
print(f'  Example: Category A → 0.42 (42% positive)')
print(f'  Pros: Reduces dimensions, captures relationship')

print('\n' + '-'*100)
print('ESTIMATED FEATURE COUNT AFTER ENCODING')
print('-'*100)

original_numerical = len(df.select_dtypes(include=['int64', 'float64']).columns)
estimated_features = original_numerical + binary_count + (lowcard_count * 3) + highcard_count

print(f'\nOriginal numerical:  {original_numerical}')
print(f'After label encode:  +{binary_count} (binary features become 1 column)')
print(f'After one-hot:       +~{lowcard_count * 3} (each creates ~3 columns)')
print(f'After target encode: +{highcard_count} (high cardinality features')
print(f'\nEstimated total:     {estimated_features} features')

print('\n' + '='*100)
print('✓ CELL 2.7 COMPLETE: ENCODING STRATEGY PLANNED')
print('='*100)
```

## CELL 2.8: SCALING STRATEGY PLANNING

### What We Do
Plan scaling strategy for numerical features.

### Scaling Methods
1. **StandardScaler**: (X - mean) / std → Mean=0, Std=1
2. **RobustScaler**: (X - median) / IQR → Robust to outliers
3. **MinMaxScaler**: (X - min) / (max - min) → Range [0, 1]

### For This Dataset
- Recommend: **StandardScaler** (most common)
- Alternative: RobustScaler (if outliers)

### Why This Matters
- Some models require normalized input
- Improves convergence speed
- Better generalization
- Standard practice in ML

```python
print('\n' + '='*100)
print('CELL 2.8: SCALING STRATEGY PLANNING')
print('='*100)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f'\nNumerical Features to Scale: {len(numerical_cols)}')
for col in numerical_cols:
    print(f'  - {col}')

print('\n' + '-'*100)
print('SCALING REQUIREMENTS ANALYSIS')
print('-'*100)

for col in numerical_cols:
    print(f'\n{col}:')
    print(f'  Range: {df[col].min():.2f} to {df[col].max():.2f}')
    print(f'  Mean: {df[col].mean():.2f}, Std: {df[col].std():.2f}')
    
    if df[col].max() > 1000 or (df[col].max() - df[col].min()) > 100:
        print(f'  Status: REQUIRES SCALING (large range)')
    else:
        print(f'  Status: Would benefit from scaling')

print('\n' + '-'*100)
print('SCALING METHODS COMPARISON')
print('-'*100)

print('\n1. StandardScaler (RECOMMENDED)')
print('   Formula: (X - mean) / std')
print('   Result: Mean=0, Std=1')
print('   Pros: Most common, works with most algorithms')
print('   Cons: Sensitive to outliers')
print('   Use when: Data approximately normally distributed')

print('\n2. RobustScaler')
print('   Formula: (X - median) / IQR')
print('   Result: Centered on median')
print('   Pros: Robust to outliers')
print('   Cons: Less interpretable')
print('   Use when: Data has significant outliers')

print('\n3. MinMaxScaler')
print('   Formula: (X - min) / (max - min)')
print('   Result: Range [0, 1]')
print('   Pros: Bounded range, interpretable')
print('   Cons: Sensitive to outliers')
print('   Use when: Bounded range is important')

print('\n' + '-'*100)
print('DECISION FOR THIS PROJECT')
print('-'*100)

print('\n✓ RECOMMENDATION: StandardScaler')
print('  Reason: Age distribution looks normal')
print('  Reason: Few outliers detected')
print('  Reason: Compatible with all models')

print('\nImplementation:')
print('  from sklearn.preprocessing import StandardScaler')
print('  scaler = StandardScaler()')
print('  X_scaled = scaler.fit_transform(X_numerical)')

print('\n' + '='*100)
print('✓ CELL 2.8 COMPLETE: SCALING STRATEGY PLANNED')
print('='*100)
```
